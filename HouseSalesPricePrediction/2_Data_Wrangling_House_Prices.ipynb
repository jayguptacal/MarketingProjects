# Data Wrangling<a id='2.0_Data_Wrangling'></a>



The focus of this Jupyter notebook is on the data wrangling steps which are as follows:

- Identification of **numerical variables** with the following analysis:     
a. Identification of **temporal variables**     
b. Identification of **discreet variables**     
c. Identification of **continuous variables**    
d. Identification of **outliers**
     
- Identification of categorical variables with the following analysis:  
a. Identification of number of labels (**cardinality**)

## Imports<a id='2.1_Imports'></a>

```python
# Importing the required packages, libraries and utilities at one place to keep it clean
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# import classes for automatic imputation techniques
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

# import classes for modeling
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split, GridSearchCV

np.random.seed(0)

# Dataframe display settings
pd.set_option('display.max_columns', None)
```

```python
# Loading the csv data file here from the data folder
data = pd.read_csv('../HousePricesPrediction/data/houseprice.csv')
```

```python
# Printing data shape and the dataset
print("Data shape is {} rows and {} columns.".format(data.shape[0], data.shape[1]))
data.head()
```

```python
# Finding types for various column features
data.info()
```

## Identification of Numerical Variables<a id='2.3_Identification_Numerical_Variables'></a>

```python
# First step is to identify the list of numerical variables and print the list

numerical_vars = [var for var in data.columns if data[var].dtypes != 'O']
print("Total number of Numerical Variables are: ", len(numerical_vars))

# Printing the dataset with the numerical variables only
data[numerical_vars].head()
```

### Identification of Temporal Variables<a id='2.3.1_Identification_Temporal_Variables'></a>

From the visual check-up, it is evident that there are following values which are of temporal type:

- **YearBuilt** - Year when the house was built
- **YearRemodAdd** - Year when the remodeling was done
- **GarageYrBlt** - Year when the garage was built
- **YrSold** - Year when the house was last sold

Based on the above information, I am going to create a list of temporal variables so that the 'year' information can be used to derive more insights.

```python
# Creating a list of temporal variables
temporal_vars = [var for var in numerical_vars if 'Year' in var or 'Yr' in var]

# Printing the unique values of the year for the temporal_vars from the data
for var in temporal_vars:
    print(var, data[var].unique())
    print()
```

### Identification of Discreet Variables<a id='2.3.2_Identification_Discreet_Variables'></a>

```python
# Finding variables which are whole numbers and part of numerical variables but not part of temporal variables
discreet_vars = [var for var in numerical_vars if len(data[var].unique()) < 20 and var not in temporal_vars+['Id']]

# Printing number of discreet variables
print("Total number of discreet variables: ", len(discreet_vars))

# Dataset for the discreet variables
data[discreet_vars].head()
```

### Identification of Continuous Variables<a id='2.3.3_Identification_Continuous_Variables'></a>

```python
# Identifying continous variables and creating a list of these variables
continuous_vars = [var for var in numerical_vars if var not in discreet_vars+temporal_vars+['Id']]
print("Total number of continuous variables are: ", len(continuous_vars))
data[continuous_vars].head()
```

### Identification of Outliers<a id='2.3.4_Identification_Outliers'></a>

```python
# Creating a function to visualize outliers for the continuous variables via box-plots
def find_outliers(df, var):
    df = df.copy()
    # calculating log values for the continous variables
    # since the log value for the negative or zero number can't be calculated, skip those values
    if any(df[var] <=0):
        pass
    else:
        df[var] = np.log(df[var])
        df.boxplot(column=var)
        plt.ylabel(var)
        plt.show()
        
for var in continuous_vars:
    find_outliers(data, var)
```

List of outliers under continuous variables is as follows:

- **Lotfrontage**
- **LotArea**
- **1stFlrSF** 
- **GrLivArea** 

These outliers will impact the performance of the machine learning algorithm, hence I will work on these outliers during the EDA phase.

## Identification of Categorical Variables<a id='2.4_Identification_Categorical_Variables'></a>

```python
# Identifying categorical variables
category_vars = [var for var in data.columns if data[var].dtypes == 'O']
# Printing number of categorical variables in the dataset
print("Total number of categorical variables are ", len(category_vars))
# dataset for the categorical variables
data[category_vars].head()
```

### Identification of Number of Labels (Cardinality)<a id='2.4.1_Number_Labels'></a>

```python
# Listing down number of unique labels for categorical variables
data[category_vars].nunique()
```

Number of labels are within the reasonable limit therefore no further action is required.

### Identification of Rare Labels<a id='2.4.2_Identification_Rare_Labels'></a>

```python
# Rare labels are those labels which are present only in a small number of houses.
# Defining a function to identify the rare labels in the categorical variables
def identify_rare_labels(df, var, rare_label_percent):
    df = df.copy()
    # Find out the percentage of observations for each category
    tmp = df.groupby(var).count()['SalePrice'] / len(df)
    # return cateogry that are rare
    return tmp[tmp<rare_label_percent]

# Categories which are less than 1% will be defined as rare labels
for var in category_vars:
    print(identify_rare_labels(data, var, 0.01))
    print()    
```

Labels that are under-represented in the dataset tend to cause over-fitting of machine learning models. Hence these will be removed during the next phase of Jupyter Notebook for Exploratory Data Analysis. 

## Missing Values Analysis<a id='2.5_Missing_Values_Analysis'></a>

```python
# Using list comprehension for building a list of variables with the missing values
vars_with_missingvalues = [var for var in data.columns if data[var].isna().sum() > 0]

# Finding out the type of missing values
data[vars_with_missingvalues].dtypes
```

From the above results, it is clear that there are missing values in both numerical as well as categorical variables.

```python
# Finding the percentage of missing values for each variable
data_na = data[vars_with_missingvalues].isna().mean()
# Transform the above array into a DataFrame
data_na = pd.DataFrame(data_na.reset_index())
# Defining columns for the missing data variables
data_na.columns = ['variable', 'missing_value_percentage']
# Sort values of the missing variables in the descending order
data_na.sort_values(by='missing_value_percentage', ascending=False, inplace=True)
# Displaying the missing values
data_na
```

### Missing Data Imputation Strategy<a id='2.5.1_MissingData_Imputation_Strategy'></a>

The following top 6 variables have lot of missing data:

- **PoolQC**
- **MiscFeature**
- **Alley**
- **Fence**
- **FireplaceQu**
- **LotFrontage**

So we can't use the simple drop or discard method for the missing data for these variables.

I am going to use **SimpleImputer** data imputation technique from the SKlearn.
- **Numerical Variable** - Mean or Median imputation
- **Categorical Variable** - Most frequent category imputation

Arbitrary value imputation is another option for both numerical and categorical variables.

I am going to create a grid search for the imputation methods to decide which imputation method will work best for these variables.

```python
# Plotting the histogram of Numerical Variables prior any imputation steps
data[numerical_vars].hist(bins=50, density=True, figsize=(12,12))
plt.tight_layout()
plt.show()
```

### Automatic Selection of Best Imputation Technique<a id='2.5.2_AutomaticSelection_BestImputation_Technique'></a>

```python
# Separating into training and testing set
X_train, X_test, y_train, y_test = train_test_split(data.drop('SalePrice', axis=1), data['SalePrice'], 
    test_size=0.3, random_state=0)
# Printing shape of both X_train and X_test to check
print("X_train shape: ", X_train.shape)
print("X_test shape: ", X_test.shape)
```

```python
features_categorical = [var for var in data.columns if data[var].dtypes=='O']
features_numerical = [var for var in data.columns if data[var].dtypes!='O' and var !='SalePrice']

# Creating pre-processing pipeline for numerical variables
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())])

# Creating pre_processing pipeline for categorical variables
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

preprocessor = ColumnTransformer(
    transformers=[
        ('numerical', numeric_transformer, features_numerical),
        ('categorical', categorical_transformer, features_categorical)])
```

```python
# Creating classifier pipeline
clf = Pipeline(steps=[('preprocessor', preprocessor),('regressor', Lasso(max_iter=1000))])
```

```python
# Creating a grid to test all the parameters
param_grid = {
    'preprocessor__numerical__imputer__strategy': ['mean', 'median'],
    'preprocessor__categorical__imputer__strategy': ['most_frequent', 'constant'],
    'regressor__alpha': [10, 100, 200],
}

# creating grid search for the cross validation
grid_search = GridSearchCV(clf, param_grid, cv=5, n_jobs=-1, scoring='r2')
```

```python
# Train all the possible combinations of the parameters above
grid_search.fit(X_train, y_train)

# Print the best score over the training set
print(("best linear regression from grid search: %.3f"
       % grid_search.score(X_train, y_train)))
```

```python
# Printing the best estimator parameters like this
grid_search.best_estimator_
```

```python
# Find the best fit parameters 
grid_search.best_params_
```

```python
# Find the scores for each of one of the above combinations
grid_search.cv_results_['mean_test_score']
```

```python
# Check the performance over the test set
print(("best linear regression from grid search: %.3f"
       % grid_search.score(X_test, y_test)))
```

For the training set, r-squared value is 0.933 and for the testing set it is 0.738. This means the model is overfitting and it needs to be tuned further.

### Finding Relationship between Missing Values and House Prices<a id='2.5.3_Finding_Missing_Values_Relationship'></a>

```python
# Defining a function to plot the relationship between the variables with the missing values and house prices.

def na_impact_prices(df, var):
    df = df.copy()
    df[var] = np.where(df[var].isna(), 1, 0) # defining '1' for missing values and '0' for no-missing values
    # plotting for the median sale prices vs. the variable with the missing values
    # Considering median and not mean prices because of missing values and outliers
    df.groupby(var)['SalePrice'].median().plot.bar(title=var + ' ' +"Missing Values Impact on Prices") 
    plt.show()
    
# Running the above function on the missing data variables
for var in vars_with_missingvalues:
    na_impact_prices(data, var)
```

Based on the missing values, it is obvious that there is some impact on the average sales price but it is hard to establish any definitive pattern.

Since the goal is to find the best model which is simple, easy to implement and easy to maintain during the production phase, another technique will be used to deal with missing values in the EDA phase keeping in mind the knowledge gained during this exercise.
