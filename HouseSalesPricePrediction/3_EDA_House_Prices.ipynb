# Exploratory Data Analysis (EDA)<a id='3.0_Exploratory_Data_Analysis'></a>

The focus of this Jupyter notebook is on the exploratory data analysis steps which are in continuation of steps done during the data wrangling phase and actions are based on the insights gained during that phase. The action steps are listed below:

- Separating Dataset into **training** and **testing** 
- Imputation of **missing values** for both numerical and categorical missing values
- Distribution of **numerical variables** to see the current distribution     
- Using **Temporal variables** to calculate the time period between 'YrSold' and other temporal variables
- Distribution of **Log Transformed numerical variables**
- Removal of **Rare Labels**     
- Encoding of **Categorical variables**
- Performing **Features scaling**
     


In summary, the focus of this step is on making the data ready for the features selection in the next step of Pre-Processing and Training Data Development.

## Importing Packages and Libraries<a id='3.1_Imports'></a>

```python
# Importing the required packages, libraries and utilities at one place to keep it clean
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import joblib

# To split data into training and testing set
from sklearn.model_selection import train_test_split

# To perform feature scaling
from sklearn.preprocessing import MinMaxScaler

# Dataframe display settings
pd.set_option('display.max_columns', None)

import warnings
warnings.simplefilter(action='ignore')
```

## Loading the Data<a id='3.2_Loading_Data'></a>

```python
# Loading the csv data file here from the data folder
data = pd.read_csv('../HousePricesPrediction/data/houseprice.csv')
```

```python
# Printing data shape and the dataset
print("Data shape is {} rows and {} columns.".format(data.shape[0], data.shape[1]))
```

## Separating Dataset into Training and Testing<a id='3.3_Separating_Dataset'></a>

In order to avoid any overfitting, I am going to separate the dataset into training and testing dataset and use the knowledge gained during the last Data Wrangling step to focus on the feature engineering steps.

```python
# Setting the seed for the random state and separating dataset into training and testing set
X_train, X_test, y_train, y_test = train_test_split(data, data['SalePrice'], test_size=0.1, random_state=0)
X_train.shape, X_test.shape
```

## Missing Values<a id='3.4_Missing_Values'></a>

### Numerical Variables Missing Values<a id='3.4.1_Numerical_Variables_MissingValues'></a>

The strategy is to impute numerical variables by following the below steps:

- Add a binary missing value indicator
- Replace the missing value with the mode

```python
# Creating a list of numerical variables of the X_train dataset
num_vars_na = [var for var in data.columns if X_train[var].isnull().sum()>0 and 
              X_train[var].dtypes != 'O']
# Print percentage of missing values for the numerical variables
X_train[num_vars_na].isnull().mean()
```

```python
# Adding a binary value and then replacing it with mode value for each numerical variable
for var in num_vars_na:
    # finding the mode value for each variable for X_train dataset
    mode_val = X_train[var].mode()[0]
    # adding binary missing indicator for both X_train and X_test
    X_train[var+'_na'] = np.where(X_train[var].isnull(), 1,0)
    X_test[var+'_na'] = np.where(X_test[var].isnull(), 1,0)
    # replacing the missing value with mode value for both X_train and X_test
    X_train[var] = X_train[var].fillna(mode_val)
    X_test[var] = X_test[var].fillna(mode_val)
    
# Checking for the missing values in both X_train and X_test
print("Numerical variables missing values in the X_train are: ", X_train[num_vars_na].isna().sum())
print()
print("Numerical variables missing values in the X_test are: ", X_test[num_vars_na].isna().sum())
```

```python
# Checking for the missing value indicators
X_train[['LotFrontage_na', 'MasVnrArea_na', 'GarageYrBlt_na']].head()
```

### Categorical Variables Missing Values<a id='3.4.2_Categorical_Variables_MissingValues'></a>

The categorical missing values will be identified and replaced with the 'Missing' value indicator.

```python
# Identify missing values list for the categorical variables
cat_vars_na = [var for var in data.columns if X_train[var].isnull().sum() >0 and X_train[var].dtypes == 'O']

# Printing missing values percentage for the categorical variables
print("Categorical missing values for X_train are: ", X_train[cat_vars_na].isnull().mean())
print()
print("Categorical missing values for X_test are: ", X_test[cat_vars_na].isnull().mean())
```

```python
# Replacing missing values for the categorical variables with 'Missing' value indicator
X_train[cat_vars_na] = X_train[cat_vars_na].fillna('Missing')
X_test[cat_vars_na] = X_test[cat_vars_na].fillna('Missing')

# Printing to check if the missing values have been replaced with 'Missing'
print("Number of missing values for X_train are: ", X_train[cat_vars_na].isnull().sum())
print()
print("Number of missing values for X_test are: ", X_test[cat_vars_na].isnull().sum())
```

## Numerical Variables<a id='3.5_Numerical_Variables'></a>

### Temporal Variables<a id='3.5.1_Temporal_Variables'></a>

During the last step of Data Wrangling, I had identified the following temporal variables:

- **YearBuilt** - Year when the house was built
- **YearRemodAdd** - Year when the remodeling was done
- **GarageYrBlt** - Year when the garage was built
- **YrSold** - Year when the house was last sold

Now I am going to focus on the elapsed time between the YrSold vs other temporal variables.

```python
# Finding temporal variables first
numerical_vars = [var for var in data.columns if data[var].dtype != 'O']
temporal_vars = [var for var in data.columns if data[var].dtype !='O' and 'Yr' or 'Year' in numerical_vars]

# defining elapsed years function
def elapsed_years(df, var):
    df[var] = df['YrSold'] - df[var]
    return df

# Calculating elpased years for X_train and X_test
for var in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    X_train = elapsed_years(X_train, var)
    X_test = elapsed_years(X_test, var)
```

### Distribution of Numerical Variables<a id='3.5.2_Distribution_Numerical_Variables'></a>

```python
# Identify the list of numerical variables to the list
num_vars = [var for var in data.columns if data[var].dtype != 'O']

# Plotting histograms for these numerical variables
def plot_histograms(df, var):
    df[var].hist(bins=30, density=True, figsize=(6,6))
    plt.tight_layout()
    plt.title(var+ "Histogram")
    plt.xlabel(var)
    plt.ylabel('Frequency')
    plt.show()

# Printing for X_train dataset
for var in num_vars:
    plot_histograms(X_train, var)
```

### Log Transformation Distribution of Numerical Variables<a id='3.5.3_Log_Transformation_Distribution'></a>

For the Gaussian distribution, I will do the log transformation of the numerical variables.

```python
# Defining the log transformation function for the numerical variables
def log_transform_plot(df, var):
    if any(df[var] <= 0):
        pass
    else:
        df[var] = np.log(df[var])
        df[var].hist(bins=30, density=True, figsize=(6,6))
        plt.title(var + "histogram")
        plt.ylabel("Frequency")
        plt.xlabel(var)
        plt.tight_layout()
        plt.show()
    
# Plotting histograms for the log-transformed numerical variables for the X_train and X_test
for var in num_vars:
    log_transform_plot(X_train, var)
```

```python
# Plotting histograms for the log-transformed numerical variables for the X_test
for var in num_vars:
    log_transform_plot(X_test, var)
```

## Removing Rare Labels<a id='3.6_Removing_Rare_Labels'></a>

I had done this step in the last Data Wrangling phase and I am now repeating it and labeling it as 'Rare'.

```python
# Defining category variables
cat_vars = [var for var in data.columns if data[var].dtype=='O']
# Finding frequent labels first
def find_frequent_labels(df, var, rare_percentage):
    df = df.copy()
    tmp = df.groupby(var)['SalePrice'].count() / len(df)
    return tmp[tmp > rare_percentage].index

# Finding rare labels and removing it from X_train & X_test 
for var in cat_vars:
    frequent_cat = find_frequent_labels(X_train, var, 0.01)
    # Replacing rare categories with 'Rare' label for both X_train and X_test
    X_train[var] = np.where(X_train[var].isin(frequent_cat), X_train[var], 'Rare')
    X_test[var] = np.where(X_test[var].isin(frequent_cat), X_test[var], 'Rare')
```

## Encoding Categorical Variables<a id='3.7_Encoding_Categorical_Variables'></a>

In this step, I will transform categorical variables strings into numerical values.

```python
# Assigning integers to categorical variables assigning smaller value to the category with lower mean house price

def replace_cat(train, test, var, target):
    ordered_labels = train.groupby([var])[target].mean().sort_values().index
    
    # Creating a dictionary of ordered labels with it's integer value
    ordinal_label = {k: i for i, k in enumerate(ordered_labels, 0)}
    
    # Using the above dictionary to replace the string values into integer values for the categories
    train[var] = train[var].map(ordinal_label)
    test[var] = test[var].map(ordinal_label)
    
# Replacing the variables with the encoded ones
for var in cat_vars:
    replace_cat(X_train, X_test, var, 'SalePrice')
```

```python
# Checking for the relationship between encoded variables and the target variable

def analyze_cat_var(df, var):
    df = df.copy()
    df.groupby(var)['SalePrice'].mean().plot.bar()
    plt.title(var)
    plt.xlabel(var)
    plt.ylabel('House Sale Price')
    plt.show()
    
# Plotting the bar chart for encoded category variables
for var in cat_vars:
    analyze_cat_var(X_train, var)
```

## Features Scaling<a id='3.8_Feature_Scaling'></a>

For the purpose of linear machine learning models, it is important that I need to normalize or scaled the features.

```python
# Capture all features in a list except for the 'Id' and 'SalePrice' columns
scaling_feats_list = [var for var in X_train.columns if var not in ['Id', 'SalePrice']]
# Counting the number of variables for scaling
print("Number of features for scaling are: ", len(scaling_feats_list))
```

```python
# Scaling with MinMaxScaler
scaler = MinMaxScaler()
# Fitting to the scaler
scaler.fit(X_train[scaling_feats_list])

# Transforming both X_train and X_test dataset
X_train[scaling_feats_list] = scaler.transform(X_train[scaling_feats_list])
X_test[scaling_feats_list] = scaler.transform(X_test[scaling_feats_list])
```

```python
# Checking the new scaled data
X_train.head()
```

```python
# Saving the train and test datasets for the next phase

X_train.to_csv('../HousePricesPrediction/data/xtrain.csv', index=False)
X_test.to_csv('../HousePricesPrediction/data/xtest.csv', index=False)
```
