# Documentation Summary For the Project <a id='6_Documentation_Summary'></a>

**Project Name: "Development of Real Estate House Prediction Demo App from the Proof of Concept (PoC) to Deployment Stage"**

## Introduction<a id='6.1_Introduction'></a>

This is a scenario-based project indicating that if I were working for a start-up company and the management team wanted me to test out a proof of concept (POC) to the deployment stage for an application which can predict the house prices based on the machine learning algorithm.

Due to the start-up environment, requirements for the project are very specific which are listed below:

- **No budget for the project** other than my time
- **No access to any paid tools**
- **No access to any data engineers** as being a start-up, I am the only data scientist on the team
- **Use only open source tools, applications, datasets and platforms**
- **Focus on the Proof of Concept while desiging the app** as it is not designed for the real world application
- **Minimum Viable Product** is the main consideration
- **Explain the project technical details in simpler terms for the benefit of everybody in the team**

## Execution Methodology<a id='6.2_Execution_Methodology'></a>

Since it is important for the project success that I keep the project simple, fast and fully documented, I adopted the following path:

- **1. Follow the Data Science Methodology (DSM)** for the machine learning model development
- **2. Create an architecture model** to keep the management team in the loop with the project developments
- **3. Create am app deployment framework** to show the deployment path clearly to various stakeholders
- **4. Follow the "STAR" methodology"** to capture the documentation and share the insights for future use

### STAR Methodology for Insights<a id='6.2.4_STAR_Methodology_Insights'></a>

Here is a summary of the **key insights** gained during the implementation of this project:

- **Use publicly available free datasets to experiment and focus on the feasibility projects**
- **It is possible to do experiments for the real-world problems without a big or no budget**
- **Focus on the speed rather than perfecting the model as these projects are designed to prove the proof of concept**
- **Use Streamlit like apps to make the machine learning model explainable and fit for presentation to various stakeholders**
- **Follow the Minimum Viability Product criteria while doing these type of projects**

#### Access House Price Prediction App on Heroku Platform<a id='6.2.4.1_Heroku_Platform_'></a>

## Future Improvements<a id='6.3_Future_Improvements'></a>

- While I have been capturing various references while doing this project, I may have missed many important references. In future I would like to create a separate place to capture all of these references so as not to miss while sharing the credits.
- In the beginning I was planning to use Flask for the web application and I invested a lot of time designing and developing the application with nice looking pages but I lost too much time in the process as I could not debug it to make it work.
- Even though there was no requirement of production coding, I missed the point and spent lot of time in completing the production coding which was not required and not used at the end.
- I used Streamlit for the development of web application and discovered many features during the process. I would like to keep working on those features for making the projects more explainable.
- Use more manageable dataset in future as this dataset was pretty cumbersome in terms of number of features and it needed more time to do the machine learning models whereas the focus was on the quick demonstration.
